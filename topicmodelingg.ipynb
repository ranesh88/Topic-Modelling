```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
#import nltk
#nltk.download('stopwords')
```

```python
stop=stopwords.words('english')
punct=list(punctuation)
bad_token=stop+punct
```

```python
def clean_text(text):
  text=text.lower()
  wordtokenize=word_tokenize(text)
  wordtokens=[t for t in wordtokenize if t.isalpha()]
  finaltokens=[t for t in wordtokens if t not in bad_token]
  return " ".join(finaltokens)
```

```python
#import nltk
#nltk.download('punkt')
```

```python
cleartext=clean_text(text)
```

```python
cleartext
```

```python
from collections import Counter
```

```python
count=dict(Counter(cleartext.split()))
```

```python
count
```

```python
import operator
sortedcount=sorted(count.items(),key=operator.itemgetter(1),reverse=True)
```

```python
sortedcount
```

```python
toptenwords=sortedcount[:10]
```

```python
toptenwords
```

```python
#Part of Speech tagging
from textblob import TextBlob
```

```python
blob=TextBlob(cleartext)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
blob=blob.tags
```

```python
blob
```

```python
blobtopten=[]
for i in toptenwords:
  for j in range(len(blob)):
    if i[0]==blob[j][0]:
      blobtopten.append(blob[j])
      break
```

```python
blobtopten
```

```python
nouns=[]
for i in blobtopten:
  if i[1]=='NN' or i[1]=='NNS' :
    nouns.append(i[0])
```
