```
import nltk
import urllib
import bs4 as bs
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
# Gettings the data source
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Global_warming').read()
```

```
# Parsing the data/ creating BeautifulSoup object
soup = bs.BeautifulSoup(source,'lxml')
```

```
# Fetching the data
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```
print(text)
```

```
# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```
text
```

```
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
nltk.download('punkt')
```

```
stop=stopwords.words('english')
punct=list(punctuation)
bad_token=stop+punct
```

```
def clean_text(text):
  wordtokenize=word_tokenize(text)
  wordtokens=[t for t in wordtokenize if t.isalpha()]
  finaltokens=[t for t in wordtokens if t not in bad_token]
  return " ".join(finaltokens)
```

```
cleartext=clean_text(text)
```

```
cleartext
```

```
from collections import Counter
```

```
count=dict(Counter(cleartext.split()))
```

```
count
```

```
import operator
sortedcount=sorted(count.items(),key=operator.itemgetter(1),reverse=True)
```

```
sortedcount
```

```
toptenwords=sortedcount[:10]
```

```
toptenwords
```

```
#Part of Speech tagging
from textblob import TextBlob
```

```
blob=TextBlob(cleartext)
```

```
import nltk
nltk.download('averaged_perceptron_tagger')
```

```
blob=blob.tags
```

```
blob
```

```
blobtopten=[]
for i in toptenwords:
  for j in range(len(blob)):
    if i[0]==blob[j][0]:
      blobtopten.append(blob[j])
      break
```

```
blobtopten
```

```
nouns=[]
for i in blobtopten:
  if i[1]=='NN' or i[1]=='NNS' :
    nouns.append(i[0])
```
